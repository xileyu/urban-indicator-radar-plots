$$
\text { Density }^{m}=\frac{|M|}{A}
$$
$$M= \text {set of occurrences of m in district, A= area of district} $$


---

Closeness to $S M=\left(\sum_{j \in B_{i}} \operatorname{dist}(j, \text { closest }(j, S M))\right)^{-1}$ 

$B_{i}=$ Set of blocks in district i $S M=$ Set of "goal" objects (Parks, Residential, Office $\ldots)$


$\operatorname{dist}(a ; b)=$ Geographical distance between two elements' centroids a and $b$ $\operatorname{closest}

(j ; Y)= Geographically closest element in set $Y$ from block j's centroid

---

\begin{aligned}
&\qquad E_{m}=\sum_{m \in M} \sum_{l \in L} v_{l}^{m} s_{l} e^{m} \\
&E_{m}=\text { Total mobility energy } \\
&M, L=\text { The set of all transport modes and road network links respectively } \\
&v_{l}^{m}=\text { The number of vehicles of mode } m \text { that travel on link } l \\
&s_{l}=\text { The length of } \operatorname{link} l \\
&e_{m}=\text { The energy usage per unit distance traveled by mode } m
\end{aligned}


---

$$
\text { Energy Efficiency }=\frac{\sum(\text { Useful Energy Output })}{\sum(\text { Energy Input })} \times 100 \%
$$
